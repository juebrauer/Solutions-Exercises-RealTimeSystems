# Introduction

This notebook is supposed to plot the results from the memory allocation / de-allocation speed test done in C.

# Get the data

We load the file using Pandas and plot it using Matplotlib.

First let us read the file.

```python
import pandas as pd

data = pd.read_csv("alloc_and_free_times.csv",
                   header=None,
                   delim_whitespace=True)
print(data)
```

```python
type(data)
```

Let us now extract the columns as normal NumPy arrays.

```python
x = data[0].values
y1 = data[1].values
y2 = data[2].values
print(type(x))
print(x)
print(y1)
print(y2)
```

# Plotting

Now we plot the allocation/deallocation times as a function of memory block size:

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12,6))

plt.xlabel("Memory block size in MB", fontsize=20)
plt.ylabel("Allocation (red)\nand deallocation (green)\ntime in ns", fontsize=20)

plt.scatter(x,y1, marker="+", color="red")
plt.scatter(x,y2, marker="o", color="green")
plt.show()
```
